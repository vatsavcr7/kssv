```python
import pandas as pd
# loading data
data=pd.read_csv(r"Downloads/salary_data.csv")

# data visualization
data.describe()
data.info()
data.isnull().sum()

X=data.iloc[:,:-1].values
y=data.iloc[:,1].values

# splitting into train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/3,random_state=0)

#linear Regression
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
lin_pred=lin_reg.predict(X_test)

# metrics and correlation
from sklearn import metrics
print('The coefficient of determination is R square = ',metrics.r2_score(y_test, lin_pred))
print('Mean squared Error = ',metrics.mean_squared_error(y_test, lin_pred))
print('Mean absolute Error = ',metrics.mean_absolute_error(y_test, lin_pred))

# Plotting graphs
import matplotlib.pyplot as plt
plt.plot(X_test,y_test)
plt.plot(X_train,y_train,color='red')
plt.plot(X_train, lin_reg.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
# Step 7 - Make new prediction
new_salary_pred = lin_reg.predict([[15]])
print("new salary prediction is:",new_salary_pred)
```
